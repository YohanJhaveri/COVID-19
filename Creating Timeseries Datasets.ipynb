## Imports

```python
import pandas as pd
import numpy as np
```

##  Datasets

```python
UPDATED_DATE = '4/21/20'

dataset = {
    'confirmed': {
        'absolute': pd.read_csv('desktop/time_series_global_confirmed.csv'),
        'scaled': pd.DataFrame(),
        'growth': pd.DataFrame(),
    },

    'deaths': {
        'absolute': pd.read_csv('desktop/time_series_global_deaths.csv'),
        'scaled': pd.DataFrame(),
        'growth': pd.DataFrame(),
    },
    
    'mortality': pd.DataFrame(),

    'daily_growth_rate': pd.DataFrame(),
}

countries_demographic = pd.DataFrame()
```

## Grouping and Removal of Redundant Columns

```python
def group(df):
    df = df.drop(columns=['Lat','Long'])
    df = pd.DataFrame(df.groupby(by = 'Country/Region').sum())
    return df

dataset['confirmed']['absolute'] = group(dataset['confirmed']['absolute'])
dataset['deaths']['absolute'] = group(dataset['deaths']['absolute'])
```

## Scaling

```python
# divide by current number of confirmed cases / deaths (max) 
# Z => [0,1]

def scaling(x):
    old_df = dataset[x]['absolute']
    new_df = pd.DataFrame()
    columns = old_df.columns
    for c in columns:
        new_df[c] = old_df[c] / old_df[UPDATED_DATE]
        dataset[x]['scaled'] = np.around(new_df, 4)
        
scaling('confirmed')
scaling('deaths')
```

## Finding Growth Rates

```python
def find_daily_growth_rate(x):
    old_df = dataset[x]['absolute']
    new_df = pd.DataFrame()
    columns = old_df.columns
    for i in range(1, len(columns)):
        old = old_df[columns[i-1]]
        new = old_df[columns[i]]
        new_df[old_df.columns[i]] = np.divide((new - old), old)
        new_df = new_df.apply(lambda x: x.fillna(0).replace(float('inf'), 0))
        dataset[x]['growth'] = np.around(new_df, 4)
```

```python
find_daily_growth_rate('confirmed')
find_daily_growth_rate('deaths')
```

```python
import numpy as np
import pandas as pd
```

```python
measures = pd.read_csv('https://raw.githubusercontent.com/YohanJhaveri/Data-Mining-Project/master/measures.csv')
```

```python
measures = measures[['COUNTRY', 'REGION', 'CATEGORY', 'MEASURE', 'TARGETED_POP_GROUP', 'DATE_IMPLEMENTED']]
```

```python
data = measures[['COUNTRY', 'MEASURE']]
```

```python
measure_to_item = {}
item_to_measure = {}

count = 1
for measure in set(data['MEASURE']):
    measure = measure.replace('\xa0','')
    measure_to_item[measure] = 'i' + str(count)
    item_to_measure['i' + str(count)] = measure
    count += 1
    
map_data = data

mapped_data = []
for measure in data['MEASURE']:
    measure = measure.replace('\xa0','')
    mapped_data.append(measure_to_item[measure])
    
map_data['MEASURE'] = mapped_data
```

```python
map_data = map_data.groupby(by="COUNTRY").agg(','.join)
map_data
```

```python
import numpy as np
import argparse
from collections import defaultdict

def apriori(input, support):

    items = {}
    freqs = {}

    def find_count(itemset):
        itemset = [items[i] for i in itemset]
        return np.sum(np.prod(data[:, itemset], axis=1))

    def delete_keys(n):
        keys = list(freqs[n].keys())
        for key in keys:
            if freqs[n][key] < support:
                del freqs[n][key]


    transactions = []
    dictionary_count = 0

    for i in input['MEASURE']:
        transaction = i.split(',')
        transactions.append(transaction)

        for item in transaction:
            if item not in items:
                items[item] = dictionary_count
                dictionary_count += 1

    data = np.zeros((len(transactions), len(items)))

    for i, transaction in enumerate(transactions):
        for item in transaction:
            data[i, items[item]] = 1

    freqs[1] = defaultdict(int)
    freqs[2] = defaultdict(int)

    for transaction in transactions:
        n = len(transaction)

        for i in range(n):
            key = (transaction[i], )
            freqs[1][key] += 1

            for j in range(i+1, n):
                key = (transaction[i], transaction[j])
                freqs[2][key] += 1

    delete_keys(1)
    delete_keys(2)

    n = 2
    
    while len(freqs[n]) > 2:
        freqs[n + 1] = {}
        keys = list(freqs[n].keys())

        for i in range(len(keys)):
            for j in range(i + 1, len(keys)):
                union = tuple(set(keys[i]).union(set(keys[j])))
                if union not in freqs[n + 1] and len(union) == (n + 1):
                    freqs[n + 1][union] = find_count(union)

        delete_keys(n + 1)
        n += 1
    for x in freqs:
        for y in freqs[x]:
            print(' '.join([item_to_measure[i] for i in y]) + ' (' + str(int(freqs[x][y])) + ')\n')

            
apriori(map_data, 5)
```
